```python
# Ans.8) A python program to iterates the integers form 1 to 50 
# showing Fizz for multiple of 3
# showing "Buzz" for multiple of 5
# showing "FizzBuzz" for multiple of 5 and 3
x=50
for i in range(1,x+1):
	if i %3==0 and i % 5==0:
		print("FizzBuzz")
	elif i % 3==0:
		print("Fizz")
	else:
		print(i)
	
  
```
